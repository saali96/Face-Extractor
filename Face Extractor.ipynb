```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import time

# Load the cascade classifier for face detection
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
def faceExtract(filePath):
    # Load the image
    img = cv2.imread(filePath)

    # Convert the image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    start_time = time.time() # start the timer
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
    end_time = time.time() # end the timer

    # Print if a face is found or not
    if len(faces) > 0:
        print("Face found!")
    else:
        print("No face found")
        return None
    # Draw rectangles around the detected faces

    for (x,y,w,h) in faces:
        cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
    # Get the cropped face
    face = img[y:y+h, x:x+w]
    # Convert the cropped face to RGB format
    face_rgb = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
    # Create a Pillow Image object from the RGB image array
    pil_face = Image.fromarray(face_rgb)

    print(f"Execution time: {end_time - start_time:.3f} seconds")

    # Show the cropped face using matplotlib
    plt.imshow(pil_face)
    plt.show()

    # Convert the original image to RGB format
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Create a Pillow Image object from the RGB image array
    pil_img = Image.fromarray(img_rgb)

    # Show the original image with rectangles around the detected faces using matplotlib
    plt.imshow(pil_img)
    plt.show()
    return pil_face
```
